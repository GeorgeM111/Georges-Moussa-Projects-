### &emsp; 1.1 Dataset Description
This dataset (**43 KB**) was obtained from **Kaggle**, a popular platform for data science competitions and datasets. You can access the dataset using the following link:  
[Dataset Link on Kaggle](https://www.kaggle.com/code/adhamtarek147/recurrence-of-thyroid-disease)

>1. Age: The age of the person in years.<br>
>2. Gender: The sex of the person, either Male (M) or Female (F).<br> 
>3. Smoking: Current smoking status; "Yes" means the person currently smokes, and "No" means they don't.<br> 
>4. Hx Smoking: A history of smoking; "Yes" means the person has smoked in the past, and "No" means they haven't.<br> 
>5. Hx Radiotherapy: Whether the person has received radiotherapy treatment in the past; "Yes" means they have, and "No" means they haven't.<br> 
>6. Thyroid Function: The current condition of the thyroid gland; can be normal (Euthyroid) or show signs of disorders like hyperthyroidism.<br> 
>7. Physical Examination: The results of a physical examination related to the thyroid, often describing the type of goiter (lumps or swelling in the thyroid).<br> 
>8. Adenopathy: The presence of enlarged lymph nodes near the thyroid; "Right" or "Left" refers to the side of the body, and "None" means no enlargement.<br> 
>9. Pathology: The type of thyroid disease found, such as different types of thyroid cancer (e.g., Micropapillary, Papillary, Hurthel cell).<br> 
>10. Focality: Describes whether the thyroid disease affects one area (Uni-Focal) or multiple areas (Multi-Focal) of the thyroid.<br> 
>11. Risk: The level of risk for the person's thyroid condition, usually categorized as "Low" or "High."<br> 
>12. T: The stage of the thyroid tumor in terms of its size and extent. For example, T1a means a small tumor, while T4b indicates an advanced stage.<br> 
>13. N: Describes whether nearby lymph nodes are affected by cancer; N0 means no affected lymph nodes, and N1b means some are affected.<br> 
>14. M: Indicates whether cancer has spread to other parts of the body; M0 means no distant spread, while M1 means it has spread.<br> 
>15. Stage: The overall stage of the cancer, combining T, N, and M, such as Stage I, Stage IVA, or IVB, which helps determine the severity.<br> 
>16. Response: The person's response to treatment. It can be "Excellent" (good outcome), "Indeterminate" (uncertain outcome), or "Incomplete" (not fully treated or controlled).<br> 
>17. Recurred: Whether the thyroid condition or cancer has come back after treatment; "Yes" means it has recurred, and "No" means it hasn't.<br>

# 5. Data Modeling

We will split the encoded dataset, standardized dataset, and discretized dataset into training and testing sets for the following reasons:

1. **Decision Tree and Random Forest:** These algorithms are not sensitive to feature scaling or data distribution. Therefore, the encoded dataset (normal dataset) is suitable for their training and testing.

2. **Gaussian Naive Bayes and K-Nearest Neighbors (KNN):** These algorithms rely on distance or distribution-based calculations. Thus, the standardized dataset is used to ensure all features contribute equally and align with Gaussian Naive Bayes assumptions.

3. **Multinomial Naive Bayes:** This algorithm is designed to work with discrete, non-negative integer features. Therefore, the discretized dataset is the most appropriate for training and testing.

```python
X = encoded_df.drop(columns='Recurred_new')
y = encoded_df['Recurred_new']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

Xs = standardized_df.drop(columns='Recurred_new')
ys = standardized_df['Recurred_new']
Xs_train, Xs_test, ys_train, ys_test = train_test_split(Xs, ys, test_size=0.3, random_state=42)

Xd = discretized_df.drop(columns='Recurred_new')
yd = discretized_df['Recurred_new']
Xd_train, Xd_test, yd_train, yd_test = train_test_split(Xd, yd, test_size=0.3, random_state=42)
```

<a id="decision"></a>

### 5.1. Decision Tree Classifier

```python
dt_param_grid = {
    'criterion': ['gini', 'entropy', 'log_loss'],
    'max_depth': [None, 5, 10, 20, 50, 100],
    'min_samples_split': [2, 5, 10, 20],
    'min_samples_leaf': [1, 2, 5, 10],
    'max_features': ['sqrt', 'log2', None],
    'class_weight': [None, 'balanced'],
    'max_leaf_nodes': [None, 10, 50, 100],
    'min_impurity_decrease': [0.0, 0.01, 0.05]
}
dt_grid_search = GridSearchCV(
    estimator=DecisionTreeClassifier(),
    param_grid=dt_param_grid,
    cv=5,
    scoring='accuracy',
    n_jobs=-1
)

dt_start_time = datetime.now()
dt_grid_search.fit(X_train, y_train)
dt_execution_time = datetime.now() - dt_start_time
```

```python
best_dt_model = dt_grid_search.best_estimator_
best_dt_params = dt_grid_search.best_params_
dt_predictions = best_dt_model.predict(X_test)

accuracy_decision = accuracy_score(y_test, dt_predictions)
precision_decision = precision_score(y_test, dt_predictions, average='weighted')
recall_decision = recall_score(y_test, dt_predictions, average='weighted')
f1_decision = f1_score(y_test, dt_predictions, average='weighted')
classification_report_dict = classification_report(y_test, dt_predictions, output_dict=True)
support_decision = classification_report_dict['weighted avg']['support']

print(f"Execution time For Decision Tree with Hyper-parameter tuning: {dt_execution_time}")
print(f"Best Parameters for Decision Tree: {dt_grid_search.best_params_}")
print("Accuracy for decision tree: ", accuracy_decision)
print("Precision for decision tree: ", precision_decision)
print("Recall for decision tree: ", recall_decision)
print("F1 Score for decision tree: ", f1_decision)
print("\nClassification Report for decision tree:")
print(classification_report(y_test, dt_predictions))
cm = confusion_matrix(y_test, dt_predictions)
print("\nConfusion Matrix for decision tree:")
print(cm)
```

<a id="random"></a>

### 5.2. Random Forest Classifier

```python
rf_param_grid = {
    'n_estimators': [50, 100, 200, 500],
    'criterion': ['gini', 'entropy'],
    'max_depth': [None, 10, 20, 50],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4],
    'max_features': ['sqrt', 'log2', None],
    'class_weight': [None, 'balanced'],
    'bootstrap': [True, False]
}
rf_grid_search = GridSearchCV(
    estimator=RandomForestClassifier(),
    param_grid=rf_param_grid,
    cv=5,
    scoring='accuracy',
    n_jobs=-1
)
rf_start_time = datetime.now()
rf_grid_search.fit(X_train, y_train)
rf_execution_time = datetime.now() - rf_start_time
```

```python
best_rf_model = rf_grid_search.best_estimator_
best_rf_params = rf_grid_search.best_params_
rf_predictions = best_rf_model.predict(X_test)

accuracy_randF = accuracy_score(y_test, rf_predictions)
precision_randF = precision_score(y_test, rf_predictions, average='weighted')
recall_randF = recall_score(y_test, rf_predictions, average='weighted')
f1_randF = f1_score(y_test, rf_predictions, average='weighted')
classification_report_rf = classification_report(y_test, rf_predictions, output_dict=True)
support_randF = classification_report_rf['weighted avg']['support']

print(f"Execution time For Random Forest with Hyper-parameter tuning: {rf_execution_time}")
print(f"Best Parameters for Random Forest: {rf_grid_search.best_params_}")
print("Accuracy for Random Forest: ", accuracy_randF)
print("Precision for Random Forest: ", precision_randF)
print("Recall for Random Forest: ", recall_randF)
print("F1 Score for Random Forest: ", f1_randF)
print("\nClassification Report for Random Forest:")
print(classification_report(y_test, rf_predictions))

cm_rf = confusion_matrix(y_test, rf_predictions)
print("\nConfusion Matrix for Random Forest:")
print(cm_rf)
```

<a id="gauss"></a>

### 5.3. Gaussian Naive Bayes

```python
gauss_NB = GaussianNB()
start_time_gauss_NB = datetime.now()
gauss_NB.fit(Xs_train, ys_train)
execution_time_gauss_NB = datetime.now() - start_time_gauss_NB
gauss_NB_predictions = gauss_NB.predict(Xs_test)
accuracy_gauss_NB = accuracy_score(ys_test, gauss_NB_predictions)
precision_gauss_NB = precision_score(ys_test, gauss_NB_predictions, average='weighted')
recall_gauss_NB = recall_score(ys_test, gauss_NB_predictions, average='weighted')
f1_gauss_NB = f1_score(ys_test, gauss_NB_predictions, average='weighted')
classification_report_gauss_NB = classification_report(ys_test, gauss_NB_predictions)
print(f"Execution Time for Gaussian Naive Bayes: {execution_time_gauss_NB}")
print("Accuracy for Gaussian Naive Bayes: ", accuracy_gauss_NB)
print("Precision for Gaussian Naive Bayes: ", precision_gauss_NB)
print("Recall for Gaussian Naive Bayes: ", recall_gauss_NB)
print("F1 Score for Gaussian Naive Bayes: ", gauss_NB)
print("\nClassification Report for Gaussian Naive Bayes:")
print(classification_report_gauss_NB)

cm_naive = confusion_matrix(y_test, gauss_NB_predictions)
print("\nConfusion Matrix for Gaussian Naive Bayes:")
print(cm_naive)
```

<a id="knn"></a>
5.5. KNearest Neighbors Classifier (KNN)

<b>We first find the best number of neighbors</b>

```python
param_grid_knn = {
    'n_neighbors': [3, 5, 10, 15, 20, 23, 30],
    'weights': ['uniform', 'distance'],                 
    'metric': ['euclidean', 'manhattan', 'minkowski', 'chebyshev'],  
    'p': [1, 2, 3],                                 
    'algorithm': ['auto', 'ball_tree', 'kd_tree', 'brute'],
    'leaf_size': [10, 20, 30, 50], 
}

knn_grid_search = GridSearchCV(
    estimator=KNeighborsClassifier(),
    param_grid=param_grid_knn,
    cv=5,
    scoring='accuracy',
    n_jobs=-1,
    verbose=2
)
start_time_knn_tuning = datetime.now()
knn_grid_search.fit(Xs_train, ys_train)
execution_time_knn_tuning = datetime.now() - start_time_knn_tuning
```

```python
best_knn_model = knn_grid_search.best_estimator_
best_knn_params = knn_grid_search.best_params_
knn_pred = best_knn_model.predict(Xs_test)

accuracy_knn = accuracy_score(ys_test, knn_pred)
precision_knn = precision_score(ys_test, knn_pred, average='weighted')
recall_knn = recall_score(ys_test, knn_pred, average='weighted')
f1_knn = f1_score(ys_test, knn_pred, average='weighted')
classification_report_knn = classification_report(ys_test, knn_pred)

print(f"Execution Time for KNN Hyperparameter Tuning: {execution_time_knn_tuning}")
print("Best Parameters for KNN:", best_knn_params)
print("Accuracy for KNN: ", accuracy_knn)
print("Precision for KNN: ", precision_knn)
print("Recall for KNN: ", recall_knn)
print("F1 Score for KNN: ", f1_knn)
print("\nClassification Report for KNN:")
print(classification_report_knn)

cm_knn = confusion_matrix(ys_test, knn_pred)
print("\nConfusion Matrix for KNN:")
print(cm_knn)
```
